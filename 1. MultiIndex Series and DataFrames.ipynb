# Multiindex series(also known as Hierarchical Indexing)
>* multiple index levels within a single index

## how to create multiindex object

>* **1. pd.MultiIndex.from_tuples()**
>* **2. pd.MultiIndex.from_product()**

>* **First way of creating multiindex DataFrame**

```python
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex = pd.MultiIndex.from_tuples(index_val)
multiindex
```

```python
multiindex.levels

# it will decouple the Multiindex data into different list
```

```python
multiindex.levels[0]
# here it will return first level of indexing 
```

```python
multiindex.levels[1]
# here it will return second level of indexing 
```

>* **Second way of creating multiindex DataFrame**

```python
pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]])
```

>* **creating a series with multiindex object**

```python
s = pd.Series([1,2,3,4,5,6,7,8],index=multiindex)
s
```

## how to fetch items from such a series

```python
s['cse']
```

```python
s['cse',2022]
```

>* **Note:** Multiindex Series is 2D

## unstack
>* it will convert the multiindex Series into DataFrame

```python
s
```

```python
s.unstack()
```

## stack
>* it will reconvert the DataFrame into multiindex Series

```python
temp = s.unstack()
temp
```

```python
temp.stack()
```

#### Then what was the point of multiindex series?
>* with help of multiindex Series and DataFrame we can represent the higher level data into 2D form such as Dataframe and Series

>* **Are columns really different from index?**
>* Pandas is bulit in such a way that it treats the index(row) and column in a same manner, so the way we have created the multiindex dataframe using rows we can do the same by giving hierarchy 

### Multiindexing in terms of columns

```python
branch_df2 = pd.DataFrame(
    [
        [1,2,0,0],
        [3,4,0,0],
        [5,6,0,0],
        [7,8,0,0],
    ],
    index = [2019,2020,2021,2022],
    columns = pd.MultiIndex.from_product([['delhi','mumbai'],['avg_package','students']])
)

branch_df2
```

```python
branch_df2['delhi']
```

```python
branch_df2['mumbai']
```

```python
branch_df2['delhi']['students']
```

```python
branch_df2.loc[2019]
# we will get the multiindex Series
```

### Multiindex df in terms of both cols and index

```python
branch_df3 = pd.DataFrame(
    [
        [1,2,0,0],
        [3,4,0,0],
        [5,6,0,0],
        [7,8,0,0],
        [9,10,0,0],
        [11,12,0,0],
        [13,14,0,0],
        [15,16,0,0],
    ],
    index = multiindex,
    columns = pd.MultiIndex.from_product([['delhi','mumbai'],['avg_package','students']])
)

branch_df3

# the given data will be of 4 dimension (4D)
```

## stacking and unstacking

```python
branch_df1
# here multiindexing is on rows in the DataFrame
```

```python
branch_df1.unstack()
# here after applying unstack, multiindexing will be on column
```

```python
'''if we apply unstack again on the above code then there will be
no row left in the DataFrame and it will be converted to another column,
and the DataFrame will be converted to Series'''

branch_df1.unstack().unstack()
```

```python
branch_df1.unstack()
```

```python
branch_df1.unstack().stack()

# it converts the inside columns into rows
```

```python
branch_df1.unstack().stack().stack()
# now it will return the Series
```

```python
branch_df2
```

```python
branch_df2.unstack()
```

```python
branch_df2.stack()
```

```python
branch_df2.stack().stack()
```

```python
branch_df3
```

```python
branch_df3.unstack()
```

```python
branch_df3.unstack().unstack()
# now it will be convertef into Series and it will be 4D Series
```

```python
branch_df3.stack()
```

```python
branch_df3.stack().stack()
```
